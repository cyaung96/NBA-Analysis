```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
pd.set_option('display.max_columns',None)
```

```python
df = pd.read_csv('2017-2018 NBA Play-By-Play Data.csv',low_memory=False)
df.head()
```

```python
final_score = df[(df['event_type'] == 'end of period')&(df['period'] == 4)]
```

```python
final_score.head()
```

```python
final_score.reset_index(inplace=True)
```

```python
# create a dictionary that has all pairs and their counts
from collections import defaultdict
dic = {}
for i in range(len(final_score)):
    a1 = final_score.loc[i,'a1']
    a2 = final_score.loc[i,'a2']
    a3 = final_score.loc[i,'a3']
    a4 = final_score.loc[i,'a4']
    a5 = final_score.loc[i,'a5']
    
    h1 = final_score.loc[i,'h1']
    h2 = final_score.loc[i,'h2']
    h3 = final_score.loc[i,'h3']
    h4 = final_score.loc[i,'h4']
    h5 = final_score.loc[i,'h5']
    
    away_pairs = [[a1,a2],[a1,a3],[a1,a4],[a1,a5],[a2,a3],[a2,a4],[a2,a5],[a3,a4],[a3,a5],[a4,a5]]
    home_pairs = [[h1,h2],[h1,h3],[h1,h4],[h1,h5],[h2,h3],[h2,h4],[h2,h5],[h3,h4],[h3,h5],[h4,h5]]
    
    for pair in away_pairs:
        
        people_1,people_2 = pair[0],pair[1]

        if people_1 + ' ' + people_2 not in dic and people_2 + ' ' + people_1 not in dic:
            dic[people_1 + ' ' + people_2] = 1
        elif people_1 + ' ' + people_2 in dic:
            dic[people_1 + ' ' + people_2] += 1
        elif people_2 + ' ' + people_1 in dic:
            dic[people_2 + ' ' + people_1] += 1
            
    for pair in home_pairs:
        
        people_1,people_2 = pair[0],pair[1]

        if people_1 + ' ' + people_2 not in dic and people_2 + ' ' + people_1 not in dic:
            dic[people_1 + ' ' + people_2] = 1
        elif people_1 + ' ' + people_2 in dic:
            dic[people_1 + ' ' + people_2] += 1
        elif people_2 + ' ' + people_1 in dic:
            dic[people_2 + ' ' + people_1] += 1

sort_dic = {k: v for k, v in sorted(dic.items(), key=lambda item: -item[1])}
sort_dic
```

```python
# check the 'Paul George' and 'Russell Westbrook' and find that they belonged to Oklahoma City in 2017.
# and find that OKC is the abbreviation for the team name.
OKC_record = final_score[final_score['GameFile'].str.contains('OKC')]
OKC_record.reset_index(inplace=True)
OKC_record.head()
```

```python
OKC_record['OKC_score'] = None
OKC_record['OKC_win'] = None
OKC_record['OKC_pair'] = None

for i in range(len(OKC_record)):
    
    away_team = OKC_record.loc[i,'GameFile'].split('@')[0][-3:]
    home_team = OKC_record.loc[i,'GameFile'].split('@')[1][0:3]
    
    if away_team == 'OKC':
        OKC_record.loc[i,'OKC_score'] = OKC_record.loc[i,'away_score']
        if OKC_record.loc[i,'away_score'] < OKC_record.loc[i,'home_score']:
            OKC_record.loc[i,'OKC_win'] = False
        else:
            OKC_record.loc[i,'OKC_win'] = True
            
        a1 = OKC_record.loc[i,'a1']
        a2 = OKC_record.loc[i,'a2']
        a3 = OKC_record.loc[i,'a3']
        a4 = OKC_record.loc[i,'a4']
        a5 = OKC_record.loc[i,'a5']
        away_list = [a1,a2,a3,a4,a5]
        
        if 'Paul George' in away_list and 'Russell Westbrook' in away_list:
            OKC_record.loc[i,'OKC_pair'] = 1
        else:
            OKC_record.loc[i,'OKC_pair'] = 0
        
        
    else:
        OKC_record.loc[i,'OKC_score'] = OKC_record.loc[i,'home_score']
        if OKC_record.loc[i,'away_score'] < OKC_record.loc[i,'home_score']:
            OKC_record.loc[i,'OKC_win'] = True
        else:
            OKC_record.loc[i,'OKC_win'] = False
        
        h1 = OKC_record.loc[i,'h1']
        h2 = OKC_record.loc[i,'h2']
        h3 = OKC_record.loc[i,'h3']
        h4 = OKC_record.loc[i,'h4']
        h5 = OKC_record.loc[i,'h5']
        home_list = [h1,h2,h3,h4,h5]
        
        if 'Paul George' in home_list and 'Russell Westbrook' in home_list:
            OKC_record.loc[i,'OKC_pair'] = 1
        else:
            OKC_record.loc[i,'OKC_pair'] = 0
```

```python
OKC_record.groupby(['OKC_pair','OKC_win']).count()['play_id']
# 0 means those 2 people did not play the game together, 1 means they did. The second column indicates if the team won.
```

```python
# hypothesis testing for difference of the average score the team gained
without_pair = OKC_record[OKC_record['OKC_pair'] == 0]
with_pair = OKC_record[OKC_record['OKC_pair'] == 1]

# check the sample size to see if it is greater than 30
print(len(without_pair),len(with_pair))
```

```python
stats.ttest_ind(a=without_pair['OKC_score'], b=with_pair['OKC_score'], equal_var=True)
```
