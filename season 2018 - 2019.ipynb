```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
pd.set_option('display.max_columns',None)
```

```python
df = pd.read_csv('2018-2019 NBA Play-By-Play Data.csv',low_memory=False)
df.head()
```

```python
final_score = df[(df['event_type'] == 'end of period')&(df['period'] == 4)]
```

```python
final_score.head()
```

```python
final_score.reset_index(inplace=True)
```

```python
final_score.to_csv('NBA Data 2018-2019',index=False)
```

```python
# create a dictionary that has all pairs and their counts
from collections import defaultdict
dic = {}
for i in range(len(final_score)):
    a1 = final_score.loc[i,'a1']
    a2 = final_score.loc[i,'a2']
    a3 = final_score.loc[i,'a3']
    a4 = final_score.loc[i,'a4']
    a5 = final_score.loc[i,'a5']
    
    h1 = final_score.loc[i,'h1']
    h2 = final_score.loc[i,'h2']
    h3 = final_score.loc[i,'h3']
    h4 = final_score.loc[i,'h4']
    h5 = final_score.loc[i,'h5']
    
    away_pairs = [[a1,a2],[a1,a3],[a1,a4],[a1,a5],[a2,a3],[a2,a4],[a2,a5],[a3,a4],[a3,a5],[a4,a5]]
    home_pairs = [[h1,h2],[h1,h3],[h1,h4],[h1,h5],[h2,h3],[h2,h4],[h2,h5],[h3,h4],[h3,h5],[h4,h5]]
    
    for pair in away_pairs:
        
        people_1,people_2 = pair[0],pair[1]

        if people_1 + ' ' + people_2 not in dic and people_2 + ' ' + people_1 not in dic:
            dic[people_1 + ' ' + people_2] = 1
        elif people_1 + ' ' + people_2 in dic:
            dic[people_1 + ' ' + people_2] += 1
        elif people_2 + ' ' + people_1 in dic:
            dic[people_2 + ' ' + people_1] += 1
            
    for pair in home_pairs:
        
        people_1,people_2 = pair[0],pair[1]

        if people_1 + ' ' + people_2 not in dic and people_2 + ' ' + people_1 not in dic:
            dic[people_1 + ' ' + people_2] = 1
        elif people_1 + ' ' + people_2 in dic:
            dic[people_1 + ' ' + people_2] += 1
        elif people_2 + ' ' + people_1 in dic:
            dic[people_2 + ' ' + people_1] += 1

sort_dic = {k: v for k, v in sorted(dic.items(), key=lambda item: -item[1])}
sort_dic
```

```python
# check the 'PJ Tucker' and 'James Harden' and find that they belonged to Houston Rockets in 2018.
# and find that HOU is the abbreviation for the team name.
HOU_record = final_score[final_score['GameFile'].str.contains('HOU')]
HOU_record.reset_index(inplace=True)
HOU_record.head()
```

```python
HOU_record['HOU_score'] = None
HOU_record['HOU_win'] = None
HOU_record['HOU_pair'] = None

for i in range(len(HOU_record)):
    
    away_team = HOU_record.loc[i,'GameFile'].split('@')[0][-3:]
    home_team = HOU_record.loc[i,'GameFile'].split('@')[1][0:3]
    
    if away_team == 'HOU':
        HOU_record.loc[i,'HOU_score'] = HOU_record.loc[i,'away_score']
        if HOU_record.loc[i,'away_score'] < HOU_record.loc[i,'home_score']:
            HOU_record.loc[i,'HOU_win'] = False
        else:
            HOU_record.loc[i,'HOU_win'] = True
            
        a1 = HOU_record.loc[i,'a1']
        a2 = HOU_record.loc[i,'a2']
        a3 = HOU_record.loc[i,'a3']
        a4 = HOU_record.loc[i,'a4']
        a5 = HOU_record.loc[i,'a5']
        away_list = [a1,a2,a3,a4,a5]
        
        if 'PJ Tucker' in away_list and 'James Harden' in away_list:
            HOU_record.loc[i,'HOU_pair'] = 1
        else:
            HOU_record.loc[i,'HOU_pair'] = 0
        
        
    else:
        HOU_record.loc[i,'HOU_score'] = HOU_record.loc[i,'home_score']
        if HOU_record.loc[i,'away_score'] < HOU_record.loc[i,'home_score']:
            HOU_record.loc[i,'HOU_win'] = True
        else:
            HOU_record.loc[i,'HOU_win'] = False
        
        h1 = HOU_record.loc[i,'h1']
        h2 = HOU_record.loc[i,'h2']
        h3 = HOU_record.loc[i,'h3']
        h4 = HOU_record.loc[i,'h4']
        h5 = HOU_record.loc[i,'h5']
        home_list = [h1,h2,h3,h4,h5]
        
        if 'PJ Tucker' in home_list and 'James Harden' in home_list:
            HOU_record.loc[i,'HOU_pair'] = 1
        else:
            HOU_record.loc[i,'HOU_pair'] = 0
```

```python
HOU_record.groupby(['HOU_pair','HOU_win']).count()['play_id']
# 0 means those 2 people did not play the game together, 1 means they did. The second column indicates if the team won.
```

```python
# hypothesis testing for difference of the average score the team gained
without_pair = HOU_record[HOU_record['HOU_pair'] == 0]
with_pair = HOU_record[HOU_record['HOU_pair'] == 1]

# check the sample size to see if it is greater than 30
print(len(without_pair),len(with_pair))
```

```python
without_pair['HOU_score'] = without_pair['HOU_score'].astype(float)
with_pair['HOU_score'] = with_pair['HOU_score'].astype(float)

stats.ttest_ind(a=without_pair['HOU_score'], b=with_pair['HOU_score'], equal_var=True)
```
